```python
class linkedNode:
    def __init__(self, val=0, next=None) -> None:
        self.next = next
        self.val = val

def linked_list_builder(nums: list) -> linkedNode:
    head = linkedNode()
    iter = head
    for num in nums:
        iter.next = linkedNode(val=num)
        iter = iter.next
    return head.next


def findKthToPast(head:linkedNode, k)->int:
    if not head:
        return [0, None] # [index, val]
    print(head.val)
    temp_result = findKthToPast(head.next, k)
    print(head.val, temp_result)
    if temp_result[1]:
        return temp_result
    index = temp_result[0] + 1
    if index == k:
        print(index)
        print("oops")
        # print(str(k) + "th to the last node is " + str(head.val))
        return [index, head.val]
    else:
        return [index, None]
    
    
    pass

k = 2
result = findKthToPast(linked_list_builder([1,2,2,3,4,5,6,7,1,3]), k)
print(result)
```

```python
def detectAndFindIntersection(head1: linkedNode, head2: linkedNode)->linkedNode:
    if not head1 or not head2:
        return None # no intersection if at least on of linked lists is empty
    len1, len2 = 1, 1
    it1, it2 = head1, head2

    while it1.next:
        it1 = it1.next
        len1 += 1
    
    while it2.next:
        it2 = it2.next
        len2 += 1

    if it1 != it2:
        return None # no intersection if the last nodes in two lists are not the same
    

    it1 = head1 if len1 > len2 else head2
    it2 = head2 if len1 > len2 else head1

    for i in range(abs(len1 - len2)):
        it1 = it1.next
    
    while it1 != it2:
        it1 = it1.next
        it2 = it2.next
    
    return it1
    
    pass
```

```python
def detectAndFindStartOfLoop(head:linkedNode) -> linkedNode:
    if not head:
        return head
    

    slow, fast = head, head
    # step1, initial the first collision
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            break
    
    # step2, put the fast to the head
    fast = head

    while fast != slow:
        fast = fast.next
        slow = slow.next

    return fast # the start point of the loop
```
