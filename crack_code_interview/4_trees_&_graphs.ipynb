```python
# 4.2 minimal Tree: build a binary tree with a sorted array and with a minimal height

class TreeNode:
    def __init__(self, val=0) -> None:
        self.left = None
        self.right = None
        self.val = val
        
        pass



def buildMinimalTree(input:list) -> TreeNode:
    # root = TreeNode()

    def TreeBuilder(l:int, r:int) -> TreeNode:
        if l > r:
            return None
        
        m = l + (r - l) // 2
        try:
            head = TreeNode(input[m])
            head.left = TreeBuilder(l, m-1)
            head.right = TreeBuilder(m+1, r)
            return head
        except IndexError:
            print(m)
    
    return TreeBuilder(0, len(input)-1)

    pass

def preorder(head:TreeNode):
    if not head:
        return
    print(head.val)
    preorder(head.left)
    preorder(head.right)

result42 = buildMinimalTree([1,2,3,4,5,6,7])
preorder(result42)
```

```python
# 4.6: Successor: write an algorithm to find the next node in an inorder traversal of 
# the given node

class TreeNode2:
    def __init__(self, val=0) -> None:
        self.left = None
        self.right = None
        self.parent = None
        self.val = val
        
        pass

def successor(node: TreeNode2) -> TreeNode2:
    if not node:
        return None
    
    if node.right:
        iter1 = node.right
        while iter1.left:
            iter1 = iter1.left
        return iter1
    else:
        if not node.parent:
            return None # node is the root and has no right tree
        else:
            if node == node.parent.left:
                return node.parent
            else:
                iter1 = node.parent
                while iter1.parent and iter1.parent.right == iter1:
                    iter1 = iter1.parent
                
                if iter1.parent: 
                    # the given node is in a left sub tree of a node
                    return iter1.parent
                else:
                    # the given node is the most right node in the tree
                    return None
    
head1 = TreeNode2(0)
head2 = TreeNode2(0)
head3 = TreeNode2(0)

it = head1
it.left = TreeNode2(1)
it.left.parent = it
it.right = TreeNode2(2)
it.right.parent = it
it = it.left
it.right = TreeNode2(3)
it.right.parent = it


it = head2
it.left = TreeNode2(1)
it.left.parent = it
it = it.left
it.left = TreeNode2(2)
it.left.parent = it
it.right = TreeNode2(3)
it.right.parent = it

it = head3
it.left = TreeNode2(1)
it.left.parent = it
it.right = TreeNode2(2)
it.right.parent = it
it = it.left
it.left = TreeNode2(3)
it.left.parent = it
it = it.parent.right
it.left = TreeNode2(4)
it.left.parent = it


node1 = head1.left.right
node2 = head2.left
node3 = head3.right.left
node4 = head3.right

print(successor(node1).val)
print(successor(node2).val)
print(successor(node3).val)
print(successor(node4))
```

```python
# 4.7: find order of building projects
import collections
def buildOrder(projects:list, dependencies:list) -> list:
    if not projects:
        return []
    if not dependencies:
        return projects
    result = []
    graph = collections.defaultdict(list)
    graph = {p:[] for p in projects}
    # 
    for dep, proj in dependencies:
        graph[proj].append(dep)
    # print(graph)
    while [] in graph.values():
        for p in projects:
            if p not in graph:
                continue
            if len(graph[p]) == 0:
                del graph[p]
                result.append(p)
                for v in graph.values():
                    if p in v:
                        v.remove(p)
            
    # print(graph)
    # print(result)
    return result if len(result) == len(projects) else []  

print(buildOrder(['a', 'b', 'c', 'd', 'e', 'f'], [['a','d'], ['f', 'b'], ['b', 'd'], ['f', 'a'], ['d', 'c']]))
print(buildOrder(['a', 'b','c'], [['b','c'], ['c','b']]))
```

```python
# 4.8:
def getSibling(node: TreeNode2):
    if not node or not node.parent:
        return None
    
    parent = node.parent
    return parent.left if node == parent.right else parent.right

def covers(root: TreeNode2, p: TreeNode2):
    if not root: 
        return False
    if root == p:
        return True
    return covers(root.left, p) or covers(root.right, p)


def FCA(root: TreeNode2, p: TreeNode2, q: TreeNode2):
    if not covers(root, p) or not covers(root, q):
        return None
    elif covers(p,q):
        return p
    elif covers(q,p):
        return q
    
    sibling = getSibling(p)
    parent = p.parent
    while not covers(sibling, q):
        sibling = getSibling(parent)
        parent = parent.parent
    return parent
```

```python
# 4.9

def weaveLists(first, second, results, prefix):
    if len(first) == 0 or len(second) == 0:
        result = prefix[:]
        result += first
        result += second
        results.append(result)
        return
    headFirst = first.pop(0)
    prefix.append(headFirst)
    weaveLists(first, second, results, prefix)
    prefix.pop()
    first.insert(0, headFirst)

    headSecond = second.pop(0)
    prefix.append(headSecond)
    weaveLists(first, second, results, prefix)
    prefix.pop()
    second.insert(0, headSecond)

def allSequence(node: TreeNode):
    results = []
    if not node:
        results.append([])
        return results
    
    prefix = []
    prefix.append(node.val)

    leftSeq = allSequence(node.left)
    rightSeq = allSequence(node.right)

    for left in leftSeq:
        for right in rightSeq:
            # print(left, right)
            weaved = []
            weaveLists(left, right, weaved, prefix)
            results += weaved
    
    return results

print(len(allSequence(result42)))
```

```python
# 4.10 check subtree (one tree is a subtree of another)

def matchTree(t1: TreeNode, t2: TreeNode):
    if not t1 or not t2:
        return True
    elif not t1 or not t2:
        return False
    elif t1.val != t2.val:
        return False
    else: # t1.val == t2.val 
        return matchTree(t1.left, t2.left) and matchTree(t1.right, t2.right) # check following subtree
    
    pass


def checkSubTree(t1: TreeNode, t2: TreeNode):
    if not t1:
        return False
    elif t1.val == t2.val and matchTree(t1, t2):
        return True
    return checkSubTree(t1.left, t2) or checkSubTree(t1.right, t2)


def containsTree(t1: TreeNode, t2: TreeNode):
    if not t2:
        return True   
    return checkSubTree(t1, t2)
```

```python
# 4.11 Random Node
import random

class TreeNode3:
    def __init__(self, val=0) -> None:
        self.val = val
        self.left = None
        self.right = None
        self.size = 1
        
        pass

    def getRandomNode(self):
        leftSize = 0 if not self.left else self.left.size
        index = random.randint(self.size)
        if index < leftSize:
            return self.left.getRandomNode()
        elif index == leftSize:
            return self
        else:
            return self.right.getRandomNode()
    
    def insertInOrder(self, d: int):
        if d < self.val:
            if not self.left:
                self.left = TreeNode3(d)
            else:
                self.left.insertInOrder(d)
        else:
            if not self.right:
                self.right = TreeNode3(d)
            else:
                self.right.insertInOrder(d)
        self.size += 1
    
    def find(self, d:int):
        if d == self.val:
            return self
        elif d <= self.val:
            return None if not self.left else self.left.find(d)
        else:
            return None if not self.right else self.right.find(d)
    
    # in a balanced tree, this algorithm will be O(log N), where N is the number of nodes
```

```python
# 4.12 path sum
def countPaths(node: TreeNode, target:int, runningSum:int, map:dict):
    if not node:
        return 0
    runningSum += node.val
    sum = runningSum - target
    totalPaths = map.get(sum, 0)
    if runningSum == target:
        totalPaths += 1
    # update map, recursion then decrease
    map[runningSum] += 1 # may use a function to check existance
    totalPaths += countPaths(node.left, target, runningSum, map)
    totalPaths += countPaths(node.right, target, runningSum, map)
    map[runningSum] -=1 # may remove the k-v pair if val is 0 to save space 
    return totalPaths
    
    pass

def countPathsWithSum(root: TreeNode, target:int):
    return countPathsWithSum(root, target, 0, dict())


# T: O(N), N = # of nodes in the tree
# as we traval to each node just once, doing O(1) work each time
# in a balanced tree, S: O(log N)
# in a unbalanced Tree, S: O(N) at most
```
