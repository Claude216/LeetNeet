```python
# 3.1 design a stack with a single array (fixed size list)
class FixedMultiStack:
    def __init__(self, numOfStacks:int, stackCapacity: int) -> None:
        self.numOfStacks = numOfStacks
        self.stackCapacity = stackCapacity
        self.values = [None for _ in range(self.stackCapacity * self.numOfStacks)] # the array
        self.sizes = [None for _ in range(numOfStacks)] # record the curr size of each stack
        pass

    # push op
    def push(self, stackNum:int, val) -> None:
        if stackNum >= self.numOfStacks or stackNum < 0:
            print("Invalid stack number, please check again!")

        if self.isFull(stackNum):
            print("Cannot push new value to this stack as it is full.")

        # get the index behind the current top 
        newIndex = self.sizes[stackNum] + stackNum * self.stackCapacity
        self.values[newIndex] = val # push the value to the stack
        self.sizes[stackNum] += 1 # update the size of this stack

    # pop op
    def pop(self, stackNum:int):
        if stackNum >= self.numOfStacks or stackNum < 0:
            print("Invalid stack number, please check again!")

        if self.isEmpty(stackNum):
            print("Cannot pop from this stack as it is empty.")

        # get the top index of the stack
        topIndex = self.sizes[stackNum] - 1 + stackNum * self.stackCapacity
        to_return = self.values[topIndex] # for return
        self.values[topIndex] = None # remove the value from the stack
        self.sizes[stackNum] -= 1 # update the size of this stack
        return to_return

    # peek op
    def peek(self, stackNum:int):
        if stackNum >= self.numOfStacks or stackNum < 0:
            print("Invalid stack number, please check again!")
        if self.isEmpty(stackNum):
            print("Cannot peek from this stack as it is empty.")
        
        # get the top index of the stack
        topIndex = self.sizes[stackNum] - 1 + stackNum * self.stackCapacity
        return self.values[topIndex]

    # isEmpty
    def isEmpty(self, stackNum:int) -> bool:
        return self.sizes[stackNum] == 0

    # isFull
    def isEmpty(self,stackNum:int) -> bool:
        return self.sizes[stackNum] == self.stackCapacity


# can also use a list to store different capacities for each stack to make it more flexible
# for different needs on the capacity
```
