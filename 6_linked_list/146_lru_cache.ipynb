```python
# class LRUCache:

#     def __init__(self, capacity: int):
#         self.max_capacity = capacity
#         self.current_cap = 0
#         self.order = []
#         self.map = {}

#     def get(self, key: int) -> int:
#         if key in self.order:
#             self.order.remove(key)
#             self.order.insert(0, key)
#             return self.map[key]
#         else:
#             return -1

#     def put(self, key: int, value: int) -> None:
#         if key in self.order:
#             self.order.remove(key)
#             self.order.insert(0, key) 
#             self.map[key] = value
#         else:
#             if self.current_cap == self.max_capacity:
#                 self.order.pop()
#                 self.order.insert(0, key)
#                 self.map[key] = value
#             else:
#                 self.order.insert(0, key)
#                 self.map[key] = value
#                 self.current_cap += 1
                

# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)



# above too slow

class Node:
    def __init__(self, key, val):
        self.key, self.val = key, val # use a linked list, each node store a key and a corresponding value
        self.prev = self.next = None # also track the previous and next nodes

class LRUCache:

    def __init__(self, capacity: int):
        self.cap = capacity # track the max cap
        self.cache = {} # use a dict to store the key-node pair
        self.left, self.right = Node(0, 0), Node(0, 0) # use a head and a tail node and track them (a double linked list)
        self.left.next, self.right.prev = self.right, self.left

    # use two helper func
    def remove(self, node):  
        prev, next = node.prev, node.next # remove a node from the linked list
        prev.next, next.prev = next, prev
        
    def insert(self, node): # insert a node to the linked list at the tail
        prev, next = self.right.prev, self.right
        prev.next = next.prev = node
        node.next, node.prev = next, prev
        
    def get(self, key: int) -> int:
        if key in self.cache:
            self.remove(self.cache[key]) # remove the key from the linked list
            self.insert(self.cache[key]) # insert the key into the linked list at the end of the list (MRU)
            return self.cache[key].val
        return -1

    def put(self, key: int, value: int) -> None:
        if key in self.cache:
            self.remove(self.cache[key]) # remove the key from the linked list (make sure it is the MRU (most recent used) after put())
        self.cache[key] = Node(key, value) # create a new node with same key and new value
        self.insert(self.cache[key]) # insert to the end of the linked list
        
        if len(self.cache) > self.cap: # if exceeds the max cap
            lru = self.left.next # get the lru
            self.remove(lru) # remove it from the linked list
            del self.cache[lru.key] # remove it and corresponding key from the cache
            
            
# idea:
# use a linked list to track the time of accessing each value
# use a dict to make it possible to access a particular node with knowing the key of it in a linked in O(1) time

            
            
```
