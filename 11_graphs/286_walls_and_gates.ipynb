```python
class Solution:
    def wallsAndGates(self, rooms: List[List[int]]) -> None:
        """
        Do not return anything, modify rooms in-place instead.
        """
        
        # idea: first, we find the gates
        #       second, we initialize a empty grid as the result to record the shortest step from the rooms to the gates
        #       second, we start from the gates, do bfs for each of the gates, and we record the steps, if the room is empty (no value in the grid
        # for this room) or contains a larger step value, we update it with new step value
        
#         ROW, COL = len(rooms), len(rooms[0])
        
        
#         gates = []
        
#         for r in range(ROW):
#             for c in range(COL):
#                 if rooms[r][c] == 0: # there is a gate
#                     gates.append((r,c))
        
        
#         def dfs(r,c, visited, step):
            
           
#             if (
#                 r < 0 or r >= ROW or
#                 c < 0 or c >= COL or
#                 (r,c) in visited or
#                 rooms[r][c] == -1
#             ):
#                 return 
            
#             if step < rooms[r][c]:
#                 rooms[r][c] = step
                
                
#             visited.add((r,c))
            
#             for n_r, n_c in [(r+1, c), (r-1, c), (r, c+1), (r, c-1)]:
#                 dfs(n_r, n_c, visited, step+1)
        
#         for r,c in gates:
#             # dfs
#             dfs(r, c, set(), 0)

        q = deque()
        visited = set()
        ROW, COL = len(rooms), len(rooms[0])
        
        def dfs(r,c):
            if (
                r < 0 or r >= ROW or
                c < 0 or c >= COL or
                (r,c) in visited or
                rooms[r][c] == -1
            ):
                return
            visited.add((r,c))
            q.append([r,c])
            
        for r in range(ROW):
            for c in range(COL):
                if rooms[r][c] == 0:
                    q.append([r,c])
                    visited.add((r,c))
        
        dist = 0
        while q:
            for i in range(len(q)): # make sure to process rooms/gates in the same step level each round
                # those meet earlier must be with a smaller step (as same step level be processed in the same round of loop)
                # so no need to compare step value on-site
                r, c = q.popleft()
                rooms[r][c] = dist
                dfs(r+1, c)
                dfs(r-1, c)
                dfs(r, c+1)
                dfs(r, c-1)
                
            dist += 1
        
        
        
        
        
```
